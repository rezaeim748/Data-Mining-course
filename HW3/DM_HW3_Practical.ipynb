# یادگیری متوالی

چندین درخت تصمیم را به صورت متوالی می سازد، جایی که هر درخت خطاهای درخت قبلی را تصحیح می کند. این به شیوه ای افزودنی کار می کند و اشتباهات پیشینیان خود را بهبود می بخشد.

# یادگیرندگان ضعیف (درخت تصمیم):

هر درخت در گروه به عنوان یک یادگیرنده ضعیف شناخته می شود. این درختان معمولاً کم عمق هستند و تنها حاوی چند گره هستند (یعنی اغلب به آنها "کاخه" می گویند).

# نزول گرادیان و باقیمانده ها:

GBM با تمرکز بر باقیمانده ها (تفاوت ها) بین مقادیر پیش بینی شده و مقادیر هدف واقعی، خطاهای ایجاد شده توسط درختان قبلی را به حداقل می رساند.
در هر تکرار، مدل گرادیان (شیب) تابع ضرر را با توجه به مقادیر پیش بینی شده محاسبه می کند. سپس درخت جدیدی را با باقیمانده‌های (خطاهای) پیش‌بینی‌های قبلی مطابقت می‌دهد.

# تقویت و وزن:

هر درخت متوالی اضافه شده به مجموعه بر کاهش خطاهای ایجاد شده توسط مدل ترکیبی (گروه درختان) تا کنون تمرکز دارد.
برای هر درخت بر اساس عملکرد آن وزن‌ها تعیین می‌شود و درختان بعدی برای توجه بیشتر به مناطقی که درختان قبلی عملکرد ضعیفی داشتند، آموزش داده می‌شوند.
